```python
% matplotlib inline
import matplotlib.pylab
import numpy as np
import pandas as pd

import pandas_datareader as web
start = pd.Timestamp('2010-1-1')
end = pd.Timestamp('2013-1-27')
f = web.DataReader("F",'yahoo',start,end)
```

```python
f.head()
```

## Plot high and low values of Yahoo's stock for July and August 2012

```python
f['2012-07':'2012-08'][['High','Low']].plot()
```

## Is the variance relatively stable over time?

```python
f.rolling(30).var()['Volume'].plot()
```

```python
f.expanding().var()['Volume'].plot()
```

## On how many days did stock close higher than it opened

```python
len (f[f.Close > f.Open])
```

## And did 'up' days become more or less frequent over time?

```python
f['DayGain'] = f.Close > f.Open
f.rolling(window=100)['DayGain'].apply(lambda x : len([x_i for x_i in x if x_i > 0])).plot()
```

```python
f.expanding()['DayGain'].apply(lambda x : len([x_i for x_i in x if x_i > 0])).plot()
```

## Compute and plot mean monthly High value of the stock

```python
f['High'].resample('M').mean().plot()
```

```python
f['High'].plot()
```

## Explore variance of the differenced volume

```python
volume = f.Volume
volume_lagged = f.Volume.diff()
diffed_volume = volume - volume_lagged
diffed_volume.rolling(window=20).var().plot()
```

## Does lagged series correlate with itself?

```python
pd.DataFrame({'read':f.Volume,'lagged':f.Volume.shift()}).corr()
```
