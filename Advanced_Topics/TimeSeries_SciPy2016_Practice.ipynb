## Windows

```python
df = pd.DataFrame(np.random.randn(10, 2), index = pd.date_range('7/1/2016', freq = 'H', periods = 10), columns = ['A', 'B'])
```

```python
%matplotlib inline 
import matplotlib.pylab
import matplotlib.pyplot as plt
```

```python
dflagged = df.shift(5)
```

```python
plt.plot(df,color='blue')
plt.plot(dflagged,color = 'red')
```

### How to lag into the past

```python
dfnegativelagged = df.shift(-5)
```

```python
plt.plot(df,color='blue')
plt.plot(dfnegativelagged,color = 'red')
```

```python
df = pd.DataFrame(np.random.randn(600, 3), index = pd.date_range('7/1/2016', freq = 'D', periods = 600), columns = ['A', 'B', 'C'])
```

```python
r = df.rolling(window=20)
```

```python
df['A'].plot(color='gray',alpha=0.5)
r.mean()['A'].plot(color='red')
```

```python
df['A'].plot(color='gray',alpha=0.5)
r.quantile(0.4)['A'].plot(color='red')
```

```python
df['A'].resample('M').mean().rolling(window=6).mean().plot()
```

# Moving Window Functions

```python
# Exponential Moving Average

ts = pd.Series(np.random.randn(100), index = pd.date_range(start='1/1/17', periods=100, freq='D'))
ts.plot()
ts.ewm(span=60,freq='D',min_periods=0,adjust=True).mean().plot(color='r')
ts.ewm(span=60,freq='D',min_periods=0,adjust=False).mean().plot()
ts.rolling(window=60).mean().plot()
```

```python
# When to use an expanding window and when to use a rolling window
```

```python
# Custom function for .quantile()
ts.rolling(window=5).apply(lambda x: sorted(x)[round(len(x)*0.5)]).plot(color = 'r')
ts.rolling(window=5).quantile(.5).plot(alpha = 0.5)
```
